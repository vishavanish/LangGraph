```python
from pydantic import BaseModel, Field
from typing import TypedDict,Annotated,Literal
import dotenv
import os
from langchain.chat_models import init_chat_model
dotenv.load_dotenv() 
```

```python
class SentimentSchema(BaseModel):
    sentiment: Literal["positive", "negative"] = Field(description="Sentiment of the Review")
```

```python
custom_profile = {
    "structured_output": True,
    # ...
}
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY","")
    
model = init_chat_model("google_genai:gemini-2.5-flash-lite", profile=custom_profile)
```

```python
structure_model = model.with_structured_output(SentimentSchema)
```

```python
prompt = """Determine if the sentiment of the following review is positive or negative - the software is very good and easy to use"""
```

```python
structure_model.invoke(prompt).sentiment
```
