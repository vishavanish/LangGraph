```python
from langgraph.graph import StateGraph , START, END
from typing import Any, TypedDict
```

```python
class BatsmanState(TypedDict):
    batsman_name: str
    runs: int
    balls: int
    fours: int
    sixes: int
    
    strike_rate: float
    balls_per_boundary: float
    boundary_percentage: float
    summary: str
```

```python
def calculate_sr(state: BatsmanState):
    sr = (state["runs"] / state["balls"]) * 100
    # return state
    ##thiss is partial update state
    return {"sr":sr}
```

```python
def calculate_bp(state: BatsmanState):
    balls_per_boundary = (state["fours"] + state["sixes"]) / state["balls"]
    return {"balls_per_boundary": balls_per_boundary}
```

```python
def calculate_boundary_percentage(state: BatsmanState):
    total_boundaries = ((state["fours"]*4) + (state["sixes"]*6))
    boundary_percentage = (total_boundaries / state["balls"]) * 100
    return {"boundary_percentage": boundary_percentage}
```

```python
def summary(state: BatsmanState):
    summary = (f"Batsman scored {state['runs']} runs off {state['balls']} balls "
                        f"with a strike rate of {state['strike_rate']:.2f}, "
                        f"balls per boundary of {state['balls_per_boundary']:.2f}, "
                        f"and a boundary percentage of {state['boundary_percentage']:.2f}%.")
    # state["summary"] = summary
    return {"summary":summary}
```

```python
graph = StateGraph(BatsmanState)
#### partial update states
graph.add_node("calculate_sr",calculate_sr)
graph.add_node("calculate_bp",calculate_bp)
graph.add_node("calculate_boundary_percentage",calculate_boundary_percentage)
graph.add_node("summary",summary)
```

```python
####  DEFINE Edges  ####
graph.add_edge(START, "calculate_sr")
graph.add_edge(START, "calculate_bp")
graph.add_edge(START, "calculate_boundary_percentage")


graph.add_edge("summary", "calculate_sr")
graph.add_edge("summary", "calculate_bp")
graph.add_edge("summary", "calculate_boundary_percentage")

graph.add_edge("calculate_sr", END)
```

```python
workflow = graph.compile()
# workflow
```

```python
intial_state = {
    "batsman_name": "Player 1",
    "runs": 120,        
    "balls": 100,
    "fours": 10,
    "sixes": 5,
    "strike_rate": 0.0,
    "balls_per_boundary": 0.0,
    "boundary_percentage": 0.0,
    "summary": ""
}
```

```python
final_workflow = workflow.invoke(intial_state)
print(final_workflow)
```
