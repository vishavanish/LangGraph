```python
from langgraph.graph import StateGraph, START,END
from typing import TypedDict
import dotenv
import os
from langchain.chat_models import init_chat_model
dotenv.load_dotenv()  # Load environment variables from .env file
os.environ["GOOGLE_API_KEY"] = ""  # Set the environment variable for Google API key
os.environ["TAVILY_API_KEY"] = ""  # Set the environment variable for Tavily API key
model = init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
model.invoke("hello")
```

```python
from langgraph.prebuilt import create_react_agent
from langchain_community.tools import TavilySearchResults
from langchain_core.tools import tool

# 1. Define your tools
@tool
def search_weather(query: str):
    """Call to search the weather."""
    return "The weather is sunny."

search_tool = TavilySearchResults(max_results=3,include_raw_content=True)
tools = [search_weather,search_tool]

agent_executor = create_react_agent(model, 
                                    tools, 
                                    verbose=True  ## real time logging of agent's thought process
                                    )

response = agent_executor.invoke({"messages": [("user", "today weather in india?")]})
print(response["messages"][-1].content)
```
